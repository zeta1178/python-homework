```python
import csv

#header row position for menu_data.csv
header_lines_menu=1
menu=[]

#header row position for sales_data.csv
header_lines_sales=1
sales=[]

#Part 1 - Step 3
with open ('menu_data.csv', 'r', newline='') as menu_csvfile:
    
    #Part 1 - Step 7
    with open ('sales_data.csv','r', newline='') as sales_csvfile:
            
        #Part 1 - Step 4
        reader_menu=csv.reader(menu_csvfile, delimiter=',')
        
        #Part 1 - Step 5
        [next(reader_menu, None) for item in range(header_lines_menu)]

        #Part 1 - Step 7
        reader_sales=csv.reader(sales_csvfile, delimiter=',')
        [next(reader_sales, None) for item in range(header_lines_sales)]

        #Part 2 - Step 1
        report={}
        
        #Part 1 - Step 6
        for row in reader_menu:
            item=(row[0])
            category=(row[1])
            description=(row[2])
            price=(row[3])
            cost=(row[4])
            menu.append([item, category, description,price,cost])
            report[row[0]] = {
                                "01-count" : 0,
                                "02-revenue": 0,
                                "03-cogs": 0,
                                "04-profit": 0,
                                "05-price": float(row[3]),
                                "06-cost": int(row[4]),
                            } 

        #Part 1 - Step 7
        for row in reader_sales:
            line_item_id=(row[0])
            date=(row[1])
            credit_card=(row[2])
            quantity=(row[3])
            menu_item=(row[4])
            sales.append([line_item_id, date, credit_card,quantity, menu_item])
            if row[4] in report:                    
                #quantity
                report[row[4]]["01-count"] += int(row[3])
                #quantity*price = revenue
                report[row[4]]["02-revenue"] += (int(row[3]))*(report[row[4]]["05-price"])
                #quantity*cost = cogs (cost?)
                report[row[4]]["03-cogs"] += (int(row[3]))*(report[row[4]]["06-cost"])
                #profit= (revenue)-(cogs) or (quantity*price)-(quantity*cost)
                report[row[4]]["04-profit"] += ((int(row[3]))*(report[row[4]]["05-price"]))-((int(row[3]))*(report[row[4]]["06-cost"]))
            else:
                print(f"{row[4]} does not equal menu_item, NO MATCH!")

        #Part 2 - Step 2
        quantity=[]
        menu_item=[]
        for s in sales:
            quantity.append([s][0][3])
            menu_item.append([s][0][4])
            
        #Part 2 - Step 3
        for m in menu_item:
            if m not in report.keys():
                report[m]={
                    "01-count" : 0,
                    "02-revenue": 0,
                    "03-cogs": 0,
                    "04-profit": 0,
                }
                    
        #Part 2 - Step 4 
        item=[]
        price=[]
        cost=[]
        for m in menu:
            item.append([m][0][2])
            price.append([m][0][3])
            cost.append([m][0][4])

#remove "05-price" key from the "report" dictionary
new_report = {a:{c:report for c, report in b.items() if c != "05-price"} for a, b in report.items()}
#remove "06-cost" key from the "new_report" dictionary
newer_report = {a:{c:new_report for c, new_report in b.items() if c != "06-cost"} for a, b in new_report.items()}
            
#remove records that have no completed sales, "01-count" = 0
#list of menu items with no sales, titled "incomplete_sale_list"
incomplete_sale_list=set([n for n in newer_report if newer_report[n]['01-count'] == 0])
#list of all possible menu items, titled "total_sale_list" 
total_sale_list=[]
for r in newer_report:
    total_sale_list.append(r)    
#list to menu items with completed sales, titled "completed_sale_list"
completed_sale_list=set(total_sale_list).difference(incomplete_sale_list) 

#Part 2 - Step 5
#send results to report file
with open('report.txt', 'w') as f:
    for c in set(completed_sale_list):
        f.write('%s,\n' % ({k:v for (k,v) in newer_report.items() if c == k}))
```
